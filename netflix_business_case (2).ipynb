Un-nesting the columns

```python
netflix['director'] = netflix['director'].apply(lambda x: x.split(', ') if isinstance(x, str) else x)
netflix['cast'] = netflix['cast'].apply(lambda x: x.split(', ') if isinstance(x, str) else x)
netflix['country'] = netflix['country'].apply(lambda x: x.split(', ') if isinstance(x, str) else x)
```

```python
netflix
```

```python

a = netflix.explode('director')[['title', 'director']].reset_index().drop(columns =['index'])


b = netflix.explode('cast')[['title', 'cast']].reset_index().drop(columns =['index'])


c = netflix.explode('country')[['title', 'country']].reset_index().drop(columns =['index'])
```

```python
a
```

```python
c
```

```python
b
```

```python

merged_ab = netflix.merge(a, on='title', how='outer')

merged_abc = merged_ab.merge(b, on='title', how='outer')

netflix_merged = merged_abc.merge(c, on='title', how='outer')
```

```python
netflix_1a = netflix_merged.drop(columns=['director_x', 'cast_x', 'country_x']).rename(columns={'director_y':'director',	'cast_y':'cast', 	'country_y':'country'})
netflix_1a
```

Treating duration columns

```python
netflix_1a['movies_duration'] = netflix_1a[netflix_1a['type']=='Movie']['duration'].apply(lambda x:int(x.split(' ')[0]) if isinstance(x, str)else x)
```

```python
netflix_1a['tv_seasons'] = netflix_1a[netflix_1a['type']=='TV Show']['duration'].apply(lambda x:int(x.split(' ')[0])if isinstance(x, str)else x)
```

```python
netflix_1a = netflix_1a.drop(columns=['duration'] )
netflix_1a
```

Treating date added column

```python

unique_dates = netflix_1a['date_added'].unique()
print(unique_dates)

netflix_1a['date_added'] = pd.to_datetime(netflix_1a['date_added'], errors='coerce', infer_datetime_format=True)

invalid_dates = netflix_1a[netflix_1a['date_added'].isnull()]['date_added']
print("Invalid Dates:")
print(invalid_dates)
```

**What is the best time to launch a TV show?**

```python
netflix_1a['week_added'] = netflix_1a['date_added'].dt.isocalendar().week
netflix_1a.head()
```

```python
tvshows_week = netflix_1a[netflix_1a['type'] == 'TV Show']
tvshows_each_week=tvshows_week.groupby('week_added')['type'].count()
tvshows_each_week.sort_values(ascending=False)
```

In the above data we can look at the top 5 weeks when the tvshows was most released

I recommend focusing on releasing new content during weeks 35, 31, 27, 53, and 24, as these periods appear to be popular for launching TV shows and movies. By aligning releases with peak weeks, Netflix can maximize viewership and engagement.



```python
movies_week = netflix_1a[netflix_1a['type'] == 'Movie']
movies_each_week=movies_week.groupby('week_added')['type'].count()
movies_each_week.sort_values(ascending=False)
```

Since most of the movies are released in the first week of every year i recommend to drop more content on the 1st week of every year

```python
week_summary = pd.merge(movies_each_week, tvshows_each_week, on='week_added', how='outer').fillna(0)
week_summary.rename(columns={'type_x' : 'movies', 'type_y':'tvshows'}).sort_values(by = 'movies', ascending = False)
```

All over i would recommend droping new content on the most popular weeks since the user will be attentive and will be expecting something new to watch

```python
netflix_1a['date_added'] = pd.to_datetime(netflix_1a['date_added'])
netflix_1a['release_year'] = pd.to_numeric(netflix_1a['release_year'], errors='coerce')

netflix_1a['days_diff'] = (netflix_1a['date_added'] - pd.to_datetime(netflix_1a['release_year'], format='%Y')).dt.days

netflix_1a.head()
```

```python
netflix_1a[['title', 'release_year', 'date_added', 'days_diff']]
```

```python
mode_days_diff = netflix_1a['days_diff'].mode()[0]

print(f"The typical time between movie release and Netflix availability is approximately {mode_days_diff} days.")
```

we see mostly there is about 2289 days gap between the release year and the date added to netflix so usually waiting for that long after the release to upload it on netflix would help the production
